## Tum ogrencilerin bilgilerini tek bir dataframe icinde su sutunlardan olusacak sekilde birlestiriniz; 
### " Sinif_Adi, Ogrenci Adi, Dogru Sayisi, Yanlis Sayisi, Bos Sayisi" .
###  Sinava giren ogrenci sayisini hesaplayiniz.
###  Siniflarin ayri ayri Dogru yanlis bos ortalamalarini hesaplayiniz.
###  Her sinifin ayri ayri en basarili ogrencisini bulunuz.
### Tum sinavin en basarili ilk 3 ogrencisini bulunuz.
###  Siniflarin standart sapmalarini hesaplayiniz. (Dogru sayilari uzerinden)
###  Sinavin dogru yanlis ve bos ortalamalarini hesaplayiniz.

```python
mind = pd.ExcelFile(r"C:\Users\Emrah\Desktop\sinif_data\py_mind.xlsx")
opinion = pd.ExcelFile(r"C:\Users\Emrah\Desktop\sinif_data\py_opinion.xlsx")
science = pd.ExcelFile(r"C:\Users\Emrah\Desktop\sinif_data\py_science.xlsx")
sense = pd.ExcelFile(r"C:\Users\Emrah\Desktop\sinif_data\py_sense.xlsx")
sinif_ismi = []
ogrenci_adi = []
dogru = []
yanlis = []
bos = []
for i in range(len(mind.sheet_names)):
    py_mind = mind.parse(i, header=None)
    sinif_ismi+=["py_mind"]
    ogrenci_adi+=[py_mind.iloc[0,0]]
    dogru+=[py_mind.iloc[-3,1]]
    yanlis+=[py_mind.iloc[-2,1]]
    bos += [py_mind.iloc[-1,1]]
    
for i in range(len(opinion.sheet_names)):
    py_opinion = opinion.parse(i, header=None)
    sinif_ismi+=["py_opinion"]
    ogrenci_adi+=[py_opinion.iloc[0,0]]
    dogru+=[py_opinion.iloc[-3,1]]
    yanlis+=[py_opinion.iloc[-2,1]]
    bos += [py_opinion.iloc[-1,1]]
    
for i in range(len(science.sheet_names)):
    py_science = science.parse(i, header=None)
    sinif_ismi+=["py_science"]
    ogrenci_adi+=[py_science.iloc[0,0]]
    dogru+=[py_science.iloc[-3,1]]
    yanlis+=[py_science.iloc[-2,1]]
    bos += [py_science.iloc[-1,1]]
    
for i in range(len(sense.sheet_names)):
    py_sense = sense.parse(i, header=None)
    sinif_ismi+=["py_sense"]
    ogrenci_adi+=[py_sense.iloc[0,0]]
    dogru+=[py_sense.iloc[-3,1]]
    yanlis+=[py_sense.iloc[-2,1]]
    bos += [py_sense.iloc[-1,1]]
    
tablo = pd.DataFrame({'Sinif Ismi': sinif_ismi,
             'Ogrenci Adi': ogrenci_adi,
             'Dogru Sayisi': dogru,
             'Yanlis Sayisi': yanlis,
             'Bos Sayisi': bos})
tablo
```

```python
tablo.Dogru Sayisi.replace('A', '0', inplace=True) # kalici degistirme
tablo

# !!! Bu degisikligi sonra duzeltecegim...
```
